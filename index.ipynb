```python
import pandas as pd

#Introducimos el link de nuestra raw
link = "https://github.com/FabioGM-02/PC2/raw/refs/heads/main/HDI.xlsx"

#Leemos el excel
hdi = pd.read_excel(link)

#En el caso de saber que filas son las que nos complican la data (Este caso son 6). Podremos utilizar de manera directa hdi = pd.read_excel(link, skiprows=6)
#Pero lo harÃ© de la manera mÃ¡s larga

#Observamos la data
display(hdi)
```

```python
#Primero eliminamos las filas que no tengan nada de data y el ranking

#Creamos la variables columnas
names = hdi.columns

#Seleccionamos las columnas que no necesitamos
drop = [names[i] for i in [3, 5, 7, 9, 11, 12, 13, 14] if i < len(names)]

#Eliminamos
hdi = hdi.drop(columns=drop)

#Cambiamos de nombres a las columnas mediante una lista
new_column_names = ["Rank", "Country", "Human_Development_Index", "Life_expectancy_at_birth", "Expected_years_of_schooling", "Mean_year_of_schooling", "Gross_national_income_per_capita"]

# Usamos el mÃ©todo `columns` del DataFrame para asignar los nuevos nombres
hdi.columns = new_column_names

display(hdi)
```

```python
#Eliminamos las primeras filas que no contienen data importante, comenzando las 5 primeras

hdi=hdi.drop(range(5))

display(hdi)
```

```python
Filasvacias = hdi[hdi['Rank'].isnull()]

#Para visualizar la informaciÃ³n puedes usar display(filas_sin_rank)

display(Filasvacias)
```

```python
#Eliminamos las filas sin rank, pero anotamos los intervalos de desarrollo humano

hdi = hdi.dropna(subset=['Rank'])

#Ahora ya no necesitamos la columna rank, la eliminamos

hdi = hdi.drop(columns=['Rank'])

display(hdi)
```

```python
#Agregamos una nueva columna Categories para anotar el desarrollo humano
hdi['Categories'] = ''

# Asignamos los rangos obtenidos de la tabla anterior
hdi.loc[7:72, 'Categories'] = 'VERY HIGH HUMAN DEVELOPMENT'
hdi.loc[74:122, 'Categories'] = 'HIGH HUMAN DEVELOPMENT'
hdi.loc[124:167, 'Categories'] = 'MEDIUM HUMAN DEVELOPMENT'
hdi.loc[169:, 'Categories'] = 'LOW HUMAN DEVELOPMENT'

#Y reiniciamos el index

hdi.reset_index(drop=True, inplace=True)

display(hdi)
```
